# Test Notebook for SDSSRM Python Module

This notebook showcases the SDSSRM (SDSS Runs Mapper) Python Module. It's capable of recording the amount of runs done throughout SDSS-observed regions 
and mapping them out onto a heatmap. Slices of the map can be done with SDSSRM.runs_map function, passing in the lower RA bound, the 
upper RA bound, the lower DEC bound, and the upper DEC bound, in that order as integers. The full map can be done by calling SDSSRM.full_map().
The module, while running, will save a .csv file of the SDSS data sample as well as a .png output of the map.

```python
import SDSSRM
```

```python
# SDSSRM.runs_map(lower RA bound, upper RA bound, lower DEC bound, upper DEC bound) 
# (bounds must be integers)
plot = SDSSRM.runs_map(0, 60, -30, 45)
```

```python
# SDSSRM.full_map() outputs the full map of runs done throughout SDSS.
fullplot = SDSSRM.full_map()
```
